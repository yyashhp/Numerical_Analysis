## Fixed point iteration

Start with 
$$
    x = g(x)
$$
and see if 
$$
    x_{k+1} = g(x_k)
$$
converges starting with a somewhat random $x_0$

Last time we saw that this works sometimes

```python
import math, numpy as np
import matplotlib.pyplot as plt
from math import pi as Ï€
```

Solve
$$
    x^5 - 3x + 1 = 0
$$

```python
f = lambda x: x**5 - 3*x + 1
```

```python
a,b = -2,2
x = np.linspace(a,b,1000)
plt.plot(x,f(x))
plt.plot([a,b],[0,0], 'k', lw = 0.5)
```

```python
a,b = 0,1.5
x = np.linspace(a,b,1000)
plt.plot(x,f(x))
plt.plot([a,b],[0,0], 'k', lw = 0.5)
```

```python
r1,r2 = 0.36, 1.22 #eyeballing from graph
plt.plot([a,b],[0,0], 'k', lw = 0.5)
plt.plot(x,f(x))
plt.plot(r1,0,'.')
plt.plot(r2,0,'.')
```

Convert to a fixed point method 
$$
    x = {x^5 + 1\over 3}
$$

```python
g = lambda x: (x**5 + 1)/3
```

```python
r3 = 1.18
plt.plot([a,b],[0,0], 'k', lw = 0.5)
plt.plot(x,g(x))
plt.plot(x,x)
plt.plot(r1,r1,'.')
plt.plot(r2,r2,'.')
plt.plot(r3,r3,'.')
plt.axis('equal')
```

```python
x0 = r2
xk = x0
for k in range(6):
    xk = g(xk)
    print(f'{k:3d} {xk:.17f}')
```

```python
x0 = r3
xk = x0
for k in range(10):
    xk = g(xk)
    print(f'{k:3d} {xk:.17f}')
```

```python
gp = lambda x: (5 * (x**4))/3
gp(xk)
```

```python
gp(xk)**10
```

## Method of Heron
To estimate $\sqrt{S}$ for $S>0$ use
$$
    x = 1 \over 2 \left({x + {S\over x}\right)
$$


```python
S = 2.
g = lambda x: (x+S/x)/2
```

```python
a,b = 0.5,2
x = np.linspace(a,b,1000)
plt.plot([a,b],[0,0], 'k', lw = 0.5)
plt.plot(x,g(x))
plt.plot(x,x)
plt.axis('equal')
```
